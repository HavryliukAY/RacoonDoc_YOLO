```python
import numpy as np
import matplotlib.pyplot as plt 
name_values = '/Users/dreamstore/Downloads/RacoonDoc/Code_2/auchan_values/labels/train/2c966368-0001_27.txt'
name_labels = '/Users/dreamstore/Downloads/RacoonDoc/Code_2/auchan_data/labels/train/2c966368-0001_27.txt'

with open(name_values) as f:
    lst_values = f.read().split('\n')
with open(name_labels) as f:
    lst_labels = f.read().split('\n')

lst_values = lst_values[:-1]
lst_labels = lst_labels[:-1]

values = np.zeros((len(lst_values),5))
labels = np.zeros((len(lst_labels),5))

for i in range(len(lst_values)):
     values[i,:] = [float(k) for k in ((''.join(str(x) for x in lst_values[i])).split())]
for i in range(len(lst_labels)):
     labels[i,:] = [float(k) for k in ((''.join(str(x) for x in lst_labels[i])).split())]
     
distances = np.zeros((len(labels), len(labels)))

for i in range(len(labels)):
    for j in range(len(values)):
        diff_x = labels[i][1] - values[j][1]
        diff_y = labels[i][2] - values[j][2]
        distances[i][j] = np.sqrt(diff_x**2 + diff_y**2)
# fig, ax = plt.subplots()
# ax.scatter(values[:, 1], values[:,2])
plt.scatter(labels[:,1], labels[:,2], label = 'Labels')
plt.scatter(values[:,1], values[:,2], color = 'orange', label = 'Values')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

n = values[:,0]


for i, txt in enumerate(n):
    plt.annotate(txt, (values[i,1], values[i,2]))
    plt.annotate(txt, (labels[i,1], labels[i,2]))
# plt.scatter(values[:, 1], values[:,2])

# plt.scatter(labels[:,1], labels[:,2], color = 'orange')
```
